```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# Correct the paths if needed
model_path = '/content/gdrive/My Drive/yolov8/runs/pose/train/weights/last.pt'
image_path = '/content/gdrive/My Drive/yolov8/samples/girl.jpg'

# Load the image
img = cv2.imread(image_path)

# Load the model
model = YOLO(model_path)

# Perform inference
results = model.predict(source=image_path, stream=True)  # get predictions

# Iterate over the results
for result in results:
    # Get the bounding boxes and keypoints
    boxes = result.boxes  # Boxes object for bounding box outputs
    keypoints = result.keypoints  # Keypoints object for keypoint outputs

    # Process detected instances
    for box, kpts in zip(boxes.xyxy.tolist(), keypoints.xy.tolist()):
        x1, y1, x2, y2 = [int(x) for x in box]  # Unpack bounding box coordinates

        # Draw bounding box
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Iterate over keypoints and draw them
        for keypoint_indx, (x, y) in enumerate(kpts):
            x, y = int(x), int(y)
            cv2.circle(img, (x, y), 5, (0, 0, 255), -1)
            cv2.putText(img, str(keypoint_indx), (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the image
    cv2_imshow(img)
    cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# Correct the paths if needed
model_path = '/content/gdrive/My Drive/yolov8/runs/pose/train/weights/last.pt'
image_path = '/content/gdrive/My Drive/yolov8/samples/man.jpg'

# Load the image
img = cv2.imread(image_path)

# Load the model
model = YOLO(model_path)

# Perform inference
results = model.predict(source=image_path, stream=True)

# Iterate over the results
for result in results:
    # Get the bounding boxes and keypoints
    boxes = result.boxes  # Boxes object for bounding box outputs
    keypoints = result.keypoints  # Keypoints object for keypoint outputs

    # Process detected instances
    for box, kpts in zip(boxes.xyxy.tolist(), keypoints.xy.tolist()):
        x1, y1, x2, y2 = [int(x) for x in box]  # Unpack bounding box coordinates

        # Draw bounding box
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Iterate over keypoints and draw them
        for x, y in kpts:
            x, y = int(x), int(y)
            cv2.circle(img, (x, y), 2, (0, 0, 255), -1)  # Reduced keypoint size

    # Display the image
    cv2_imshow(img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
